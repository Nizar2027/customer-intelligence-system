```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/online_retail.csv",
                 parse_dates=["InvoiceDate"])

df.head()
```

```python
CHURN_DAYS = 90
```

```python
max_date = df["InvoiceDate"].max()
cutoff_date = max_date - pd.Timedelta(days=CHURN_DAYS)

print("Max date:", max_date)
print("Cutoff date:", cutoff_date)
```

```python
train_df=df[df["InvoiceDate"]<=cutoff_date]
test_df=df[df["InvoiceDate"]>cutoff_date]

print("train transactions",len(train_df))
print("test transactions",len(test_df))
```

```python
train_df.to_csv("../data/processed/train_transactions.csv",index=False)
test_df.to_csv("../data/processed/future_transactions.csv",index=False)
```
